### Team Shooting

```python
df_team_shooting_1 = pd.read_csv('data\Raw Data\Team Stat Shooting Team v1.csv')
df_team_shooting_2 = pd.read_csv('data\Raw Data\Team Stat Shooting Team v2.csv')
```

```python
df_team_shooting_2.head(3)
```

Remove all blank columns due to table format shown in basketball reference

```python
blank_cols = [col for col in df_team_shooting_1 if "\xa0" in col]
df_team_shooting_1.drop(columns= blank_cols, inplace = True)
```

```python
blank_cols = [col for col in df_team_shooting_2 if "\xa0" in col]
df_team_shooting_2.drop(columns= blank_cols, inplace = True)
```

```python
df_team_shooting_2.head(5)
```

Remove layup stats from v2 shooting table as there were not stats for anything prior to 2002

#### Fixing naming convention for v1 shooting table (2000 - 2001)

```python
col_range = range(df_team_shooting_1.columns.get_loc('2P'), df_team_shooting_1.columns.get_loc('3P') + 1)
columns_FGA_percent = df_team_shooting_1.iloc[:,col_range]

df_team_shooting_1.rename(columns = {col: 'per_FGA_' + col for col in columns_FGA_percent}, inplace = True)
```

```python
col_range = range(df_team_shooting_1.columns.get_loc('2P.1'), df_team_shooting_1.columns.get_loc('3P.1') + 1)
columns_FG = df_team_shooting_1.iloc[:,col_range]

df_team_shooting_1.rename(columns = {col: 'FGA%_' + col.replace('.1','') for col in columns_FG}, inplace = True)
```

```python
col_range = range(df_team_shooting_1.columns.get_loc('2P.2'), df_team_shooting_1.columns.get_loc('3P.2') + 1)
columns_FG_astd = df_team_shooting_1.iloc[:,col_range]

df_team_shooting_1.rename(columns = {col: "FG_astd_" + col.replace('.2','') for col in columns_FG_astd}, inplace = True)
```

```python
col_range = range(df_team_shooting_1.columns.get_loc('%FGA'), df_team_shooting_1.columns.get_loc('Md.') + 1)
columns_dunk = df_team_shooting_1.iloc[:,col_range]

df_team_shooting_1.rename(columns = {col: 'dunk_' + col for col in columns_dunk}, inplace = True)
```

```python
col_range = range(df_team_shooting_1.columns.get_loc('%3PA'), df_team_shooting_1.columns.get_loc('3P%') + 1)
columns_corner = df_team_shooting_1.iloc[:,col_range]

df_team_shooting_1.rename(columns = {col: 'corner_' + col for col in columns_corner}, inplace = True)
```

```python
col_range = range(df_team_shooting_1.columns.get_loc('Att.'), df_team_shooting_1.columns.get_loc('Md..1') + 1)
columns_heaves = df_team_shooting_1.iloc[:,col_range]

df_team_shooting_1.rename(columns = {col: 'heaves_' + col.replace(".2","") for col in columns_heaves}, inplace = True)
```

#### Fixing naming convention for v2 shooting table (2002 - 2024)

```python
df_team_shooting_2.drop(columns = ['%FGA.1', 'Md..1'], inplace = True)
```

```python
col_range = range(df_team_shooting_2.columns.get_loc('2P'), df_team_shooting_2.columns.get_loc('3P') + 1)
columns_FGA_percent = df_team_shooting_2.iloc[:,col_range]

df_team_shooting_2.rename(columns = {col: 'per_FGA_' + col for col in columns_FGA_percent}, inplace = True)
```

```python
col_range = range(df_team_shooting_2.columns.get_loc('2P.1'), df_team_shooting_2.columns.get_loc('3P.1') + 1)
columns_FG = df_team_shooting_2.iloc[:,col_range]

df_team_shooting_2.rename(columns = {col: 'FGA%_' + col.replace('.1','') for col in columns_FG}, inplace = True)
```

```python
col_range = range(df_team_shooting_2.columns.get_loc('2P.2'), df_team_shooting_2.columns.get_loc('3P.2') + 1)
columns_FG_astd = df_team_shooting_2.iloc[:,col_range]

df_team_shooting_2.rename(columns = {col: "FG_astd_" + col.replace('.2','') for col in columns_FG_astd}, inplace = True)
```

```python
col_range = range(df_team_shooting_2.columns.get_loc('%FGA'), df_team_shooting_2.columns.get_loc('Md.') + 1)
columns_dunk = df_team_shooting_2.iloc[:,col_range]

df_team_shooting_2.rename(columns = {col: 'dunk_' + col for col in columns_dunk}, inplace = True)
```

```python
col_range = range(df_team_shooting_2.columns.get_loc('%3PA'), df_team_shooting_2.columns.get_loc('3P%') + 1)
columns_corner = df_team_shooting_2.iloc[:,col_range]

df_team_shooting_2.rename(columns = {col: 'corner_' + col for col in columns_corner}, inplace = True)
```

```python
col_range = range(df_team_shooting_2.columns.get_loc('Att.'), df_team_shooting_2.columns.get_loc('Md..2') + 1)
columns_heaves = df_team_shooting_2.iloc[:,col_range]

df_team_shooting_2.rename(columns = {col: 'heaves_' + col.replace(".2","") for col in columns_heaves}, inplace = True)
```

```python
blank_columns = df_player_shooting.columns[df_player_shooting.columns.str.contains('\xa0')]
df_player_shooting.drop(columns = blank_columns, inplace = True)
```

```python
df_player_shooting.drop(0, axis = 0, inplace = True)
```

```python
df_player_shooting.columns
```

```python
columns_percent_FGA = range(df_player_shooting.columns.get_loc('2P'), df_player_shooting.columns.get_loc('3P') + 1)
for num in columns_percent_FGA:
    col = df_player_shooting.columns[num]
    df_player_shooting.rename(columns = {col: f'%_FGA_{col}'}, inplace = True)
```

```python
columns_FG_percent = range(df_player_shooting.columns.get_loc('2P.1'), df_player_shooting.columns.get_loc('3P.1') + 1)
for num in columns_FG_percent:
    col = df_player_shooting.columns[num]
    col_rename = df_player_shooting.columns[num].replace('.1','')
    df_player_shooting.rename(columns = {col: f'FGA%_{col_rename}'}, inplace = True)
```

```python
df_player_shooting.rename(columns = {'2P.2': '%_FG_ast_2P', '3P.2': '%_FG_ast_3P', '%FGA': '%_FGA_dunks', '#': "#_dunks", 
                                      '%3PA': 'corner_3s_%3PA', '3P%': 'corner_3s_3P%', 'Att.': 'heaves_atts', '#.1': '#_heaves'}, inplace = True)
```
